```python
def rgb_to_grayscale(tensor):
    weights = torch.tensor([0.2989, 0.5870, 0.1140]).view(1, 3, 1, 1).to(tensor.device)
    grayscale_tensor = (tensor * weights).sum(dim=1, keepdim=True)
    return grayscale_tensor
```
